```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.ylim(0,100)
x = np.arange(1,10,.1)
y_constant = np.zeros((len(x),)) + 20
y_linear = x
y_n2 = x ** 2
y_nlogn = x*np.log(x)
plt.xlabel("input size")
plt.ylabel("execution time / resources")
plt.plot(x, y_constant, label="O(1)")
plt.plot(x, y_linear, label="O(n)")
plt.plot(x, y_n2, label="O(n2)")
plt.plot(x, y_nlogn, label="O(nlog(n))")

plt.legend()
```
